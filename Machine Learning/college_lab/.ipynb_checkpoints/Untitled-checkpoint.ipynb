```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans 
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
# Load the iris dataset (as an example)
iris = load_iris()
iris_df = pd.DataFrame(data= np.c_[iris['data'], iris['target']], columns= iris['feature_names'] + ['target'])
```

```python
# Save the iris dataset to a CSV file
iris_df.to_csv('iris_dataset.csv', index=False)

# Load the dataset from the CSV file
iris_df = pd.read_csv('iris_dataset.csv')
```

```python
# Selecting only two features for simplicity (you can choose different features based on your dataset)
X = iris_df[['sepal length (cm)', 'sepal width (cm)']]

# Split the data into training and testing sets
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)
```

```python
# Apply K-means clustering on the training data
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_train)
```

```python
# Get the centroids and labels
centroids = kmeans.cluster_centers_
labels_train = kmeans.labels_
```

```python
# Visualize the clustered training data with centroids
plt.scatter(X_train['sepal length (cm)'], X_train['sepal width (cm)'], c=labels_train, s=50, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, linewidths=3, color='r', label='Centroids')
plt.title('K-means Clustering of Iris Training Data')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.show()
```

```python
# Predict the cluster for the testing data
labels_test = kmeans.predict(X_test)
```

```python
# New data points for prediction
new_data_points = np.array([[5.5, 3.2], [6.0, 3.0], [4.5, 2.5]])
```

```python
# Predict the cluster for the new data points
predicted_labels_new_data = kmeans.predict(new_data_points)
```

```python
# Visualize the clustered testing data with centroids and the new data points
plt.scatter(X_test['sepal length (cm)'], X_test['sepal width (cm)'], c=labels_test, s=50, cmap='viridis', label='Testing Data')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, linewidths=3, color='r', label='Centroids')
plt.scatter(new_data_points[:, 0], new_data_points[:, 1], c=predicted_labels_new_data, marker='D', s=100, cmap='viridis', label='New Data Points')
plt.title('K-means Clustering of Iris Testing Data with New Data Points')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.show()
```
