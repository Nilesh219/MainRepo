```python
import pandas as pd

df = pd.read_csv("Product_data.csv")
unique_products = df.drop_duplicates()  # Remove duplicate rows
total_unique_products = len(unique_products)

print("Total number of unique products:", total_unique_products)
```

```python
import csv

# Assuming "product_code" is a column in the CSV
product_code_counts = {}
with open("Product_data.csv") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        code = row["productid"]
        product_code_counts[code] = product_code_counts.get(code, 0) + 1

# Print total products per category (product code)
for code, count in product_code_counts.items():
    print(f"Product code: {code}, Total products: {count}")
```

```python
import csv

product_category_counts = {}
with open("Product_data.csv") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        category = row["productname"]
        code = row["productid"]
        product_category_counts[category] = product_category_counts.get(category, {})
        product_category_counts[category][code] = product_category_counts[category].get(code, 0) + 1

# Print total products per category (product category)
for category, code_counts in product_category_counts.items():
    print(f"Category: {category}")
    for code, count in code_counts.items():
        print(f"\tProduct code: {code}, Total products: {count}")
```
