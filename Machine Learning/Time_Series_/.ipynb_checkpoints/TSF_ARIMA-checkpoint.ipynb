```python
import itertools
p = d = q = range(0,2)

pdq = list(itertools.product(p, d, q))

seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p,d,q))]

print('Examples of parameter combinations for seasonal ARIMA..')
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[2]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[3]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[4]))
```

```python
pdq
```

```python
seasonal_pdq
```

```python
import statsmodels.api as sm
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            model = sm.tsa.statespace.SARIMAX(new_df,
                                             order=param,
                                             seasonal_order=param_seasonal,
                                             enforce_stationarity=False,
                                             enforce_invertibility=False)
            results = model.fit()
            print(f'ARIMA{param}x{param_seasonal}12 - AIC:{results.aic}')
        except:
            continue
```

```python
Best_model = sm.tsa.statespace.SARIMAX(new_df,
                                      order=(1,1,1),
                                      seasonal_order=(1,1,1,12),
                                      enforce_stationarity=True,
                                      enforce_invertibility=False)

results = Best_model.fit()

print(results.summary().tables[1])
```

```python
results.plot_diagnostics(figsize=(15,12))
plt.show()
```

```python
#Confidence interval
pred = results.get_prediction(start=pd.to_datetime('2023-03-22 00:00:00'), dynamic=False)
pred_ci = pred.conf_int()
pred_ci
```

```python
#Actual and Predicted Values
y_forecasted = pred.predicted_mean
y_truth = new_df['2023-03-22 00:00:00' :]
y_truth['pred_quantity'] = y_forecasted
y_truth
```

```python
#visualize actual and predicted values
ax = new_df['2023-03-22 00:00:00':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahed Forecast', alpha=.7)

ax.fill_between(pred_ci.index,
               pred_ci.iloc[:, 0],
               pred_ci.iloc[:, 1], color='k', alpha=.2)
ax.set_xlabel('Date')
ax.set_ylabel('quantity')
plt.legend()

plt.show()
```

```python
#compute  the mean error square error
mse = ((y_truth['quantity'] - y_truth.pred_quantity)**2).mean()
print('Mean Squraed Error = ',round (mse,2))
```

```python
rmse = np.sqrt(mse)
print('Mean Squared Error = ', rmse)
```

```python
# Testing for Stationarity
from statsmodels.tsa.stattools import adfuller
```

```python
test_result = adfuller(new_df['quantity'])
```

```python
print(new_df['quantity'].describe())
```

```python
# H0: It is non stationary
# H1: It is stationary

def adfuller_test(quantity):
    result = adfuller(quantity)
    labels = ['ADF Test Statistic', 'p-value', '#Lags Used', 'Number of observations Uesd']
    for value,label in zip(result,labels):
        print(label+' : '+str(value))
        if result[1] <= 0.05:
            print("Strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data has no unit root and is stationary")
        else:
            print("weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary")
```

```python
adfuller_test(new_df['quantity'])
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
# Identify seasonality (e.g., monthly, quarterly) based on your data
seasonality = 12  # Adjust if seasonality is different (e.g., 4 for quarters)

# Define the SARIMA model with initial parameter values (adjust as needed)
model = SARIMAX(
    train['quantity'],
    order=(2, 1, 0),  # ARIMA order (p, d, q)
    seasonal_order=(1, 1, 1, seasonality),  # Seasonal ARIMA order (p, d, q, S)
)

# Fit the model to the training data
model_fit = model.fit()

# Print the model summary (optional)
print(model_fit.summary())

# Generate forecasts for the test set (or future periods)
forecast_values = model_fit.forecast(steps=len(test))
```

```python
print(forecast_values)
```
